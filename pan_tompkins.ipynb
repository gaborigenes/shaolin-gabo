```python
import wfdb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

archivo = f'/content/physionet.org/files/afpdb/1.0.0/p01c'
registro = wfdb.rdrecord(archivo, sampfrom=180, sampto=4000,)    
anotacion = wfdb.rdann(archivo, 'dat', sampfrom=180, sampto=4000,shift_samps=True)
wfdb.plot_wfdb(record=registro, annotation=anotacion, time_units='seconds',figsize=(20,10))
```
